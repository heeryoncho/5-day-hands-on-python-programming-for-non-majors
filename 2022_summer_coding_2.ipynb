### **2. 튜플** (https://wikidocs.net/15)

튜플(tuple)은 몇 가지 점을 제외하고 리스트와 거의 비슷합니다. 

튜플이 리스트와 다른 점은 다음과 같습니다.

- 리스트는 [ ] 로 표현하지만 **튜플은 ( ) 로 표현**합니다.
- 리스트는 요소들의 생성, 삭제, 수정이 가능하지만, **튜플은 한 번 생성된 요소를 바꿀 수 없습니다.**

튜플의 예시는 다음과 같습니다.

```
t1 = ()    # 빈 튜플
t2 = (1,)    # 한 개의 요소만 있을 경우, 뒤에 반드시 쉼표(,)를 붙여야 함
t3 = (1, 2, 3)
t4 = 1, 2, 3    # 괄호를 생략해도 됨
t5 = ('a', 'b', ('ab', 'cd'))
```

프로그램이 실행되는 동안 요소들이 고정되기를 원할 경우 튜플을 사용하는 것이 좋습니다.

실제 프로그램에서는 값이 변경되는 경우가 많기 때문에 보통 튜플보다는 리스트를 더 많이 사용합니다.


---
실제로 튜플의 요소를 삭제할 수 있는지 실험해 봅시다.

````python
t3 = (1, 2, 3)
# del t3[1]    # 에러가 발생합니다.
````

---
튜플은 값을 변화시킬 수 없다는 점만 제외하고, 리스트에서 사용했던 다양한 함수들을 그대로 사용할 수 있습니다.

아래에 **인덱싱, 슬라이싱, 튜플 더하기, 튜플 곱하기, 튜플 길이 구하기**에 대한 예를 제시합니다.

### **3. 집합** (https://wikidocs.net/1015)

**집합(set)** 자료형은 **중복 요소가 없는 고유한 요소들의** 집합을 만들 수 있는 자료형입니다.

집합을 이용하면 **교집합, 합집합, 차집합 등의 집합 연산**이 가능합니다.

집합이 리스트나 튜플과 다른 점은 다음과 같습니다.

- 중복을 허용하지 않습니다.
- 순서가 없습니다. (unordered)

리스트나 튜플은 순서가 있기(ordered) 때문에 인덱싱을 통해 요소를 선택할 수 있었지만,

**집합 자료형은 순서가 없기 때문에 인덱싱으로 요소를 선택할 수 없습니다.**

```
s1 = set( )   # 빈 집합
s2 = {1, 2, 3}
s3 = set(["apple", "banana", "cherry"])
s4 = {True, -32, 'Python'}
s5 = set(('a', 'b', 'c'))
```

집합은 요소를 추가하거나 삭제할 수 있기 때문에 가변적입니다.

또 집합은 중복을 허용하지 않기 때문에 **리스트 속 요소들의 중복을 제거하기 위한 필터**로 사용됩니다.

(예: 단어 리스트에서 고유단어 취득.)

````python
s3 = set(["apple", "banana", "banana", "cherry"])
s3
````

````python
s5 = set(('a', 'b', 'c', 'a'))
s5
````

---
3-1. **교집합(intersection)** 구하기

두 집합의 교집합을 **'&' 기호** 와 **intersection( )** 함수의 두 가지 방법으로 구할 수 있습니다.

````python
s1 = {1, 2, 3, 4, 5, 6}
s2 = {4, 5, 6, 7, 8, 9}
s3 = {1, 3, 4, 6, 9, 0}
````

````python
# '&' 기호를 이용하는 방법

print(s1 & s2)
print(s1 & s2 & s3)
````

````python
# intersection() 함수를 이용하는 방법

print(s1.intersection(s2))
print(s1.intersection(s2, s3))
````

---
3-2. **합집합(union)** 구하기

두 집합의 합집합은 **'|' 기호**와 **union( )** 함수의 두 가지 방법으로 구할 수 있습니다.

````python
# '|' 기호를 이용하는 방법

print(s1 | s2)
print(s1 | s2 | s3)
````

````python
# union() 함수를 이용하는 방법

print(s1.union(s2))
print(s1.union(s2, s3))
````

---
3-3. **차집합(difference)** 구하기

두 집합의 차집합은 **'-' 기호** 와 difference( ) 함수의 두 가지 방법으로 구할 수 있습니다.

````python
s1 = {1, 2, 3, 4, 5, 6}
s2 = {4, 5, 6, 7, 8, 9}
````

````python
# '-' 기호를 이용하는 방법

print(s1 - s2)
print(s2 - s1)
````

````python
# difference() 함수를 이용하는 방법

print(s1.difference(s2))
print(s2.difference(s1))
````

### **4. 딕셔너리**(https://wikidocs.net/16)

일반적으로 단어의 뜻을 정의한 사전은 등록어와 등록어에 대한 설명으로 구성됩니다.

```
사과 : 붉은 색 과일
참외 : 노란 색 과일
```
이와 비슷한 형태로 데이터를 저장할 수 있는 자료형이 **딕셔너리(dictionary)** 자료형입니다.

```
dic = { }   # 빈 딕셔너리
person = {'name':'홍길동', 
          'mobile':'010-9999-0000', 
          'point':3200, 
          'rating':'silver'}
```

```
characters = {'DC':['Batman', 'Superman', 'Wonder Woman'],
              'Marvel':['Spider-Man', 'Iron Man', 'Captain America']}
```
이와 같이 딕셔너리 자료형은 **key** 와 **value** 가 한 쌍을 이루는 자료형입니다.

위의 person 과 character 딕셔너리에서 key 와 value 는 각각 아래와 같습니다. 

- key - 'name', 'mobile', 'point', 'rating', 'DC', 'Marvel'
- value - '홍길동', '010-9999-0000', 3200, 'silver', ['Batman', ...], ['Spider-Man', ...]

딕셔너리는 아래와 같이 key 와 value 의 쌍을 정의하는 것으로 만들 수 있습니다.

```
{key1: value1, key2: value2, key3: value3, ...}
```

여기서 주의해야 할 점은 **key가 중복되면 안된다**는 것입니다.

또 key 에 리스트 자료형은 사용할 수 없습니다. (숫자, 문자열, 튜플 등 고정된 자료형만 사용 가능.)


````python
heroes = {'Batman': 'DC', 'Spider-Man': 'Marvel', 'Hulk': 'Marvel'}
heroes
````

### **5. 리스트(list) ↔ 집합(set) 변환**

리스트를 집합으로, 집합을 리스트로 변환할 수 있습니다.

````python
some_list = [1, 2, 3, 4, 5]

print(some_list)
print(type(some_list), '\n')

some_set = set(some_list)   # set() 함수를 이용하여 집합으로 변환

print(some_set)
print(type(some_set))
````

````python
repetitive_item_list = [1, 1, 1]

set(repetitive_item_list)   # set() 함수로 리스트의 중복을 제거
````

````python
some_set = {'a', 'b', 'c'}

print(some_set)
print(type(some_set), '\n')

some_list = list(some_set)   # list() 함수를 이용하여 리스트로 변환

print(some_list)
print(type(some_list))
````
